```python
import pandas as pd
from pathlib import Path
import re
```

```python
#read the csv from the google API specifically the epidemiology csv
epidem = pd.read_csv("https://storage.googleapis.com/covid19-open-data/v3/epidemiology.csv")
```

```python
#showcase the data
epidem.head()
```

```python
#double check the datatypes of the import
epidem.info()
```

```python
#limit the dates to those of the first infection reported
epidem_group = epidem.loc[epidem['cumulative_confirmed']>=1]
epidem_group
```

```python
#dropping any rows that have a NA in the location key
epidem_group = epidem_group.dropna(subset=['location_key'])
epidem_group
```

```python
#we only want US data so by using RegEx we are able to filter out the US locations
pattern = r'^US_..$'

us_epi = epidem_group.loc[epidem_group['location_key'].str.contains(pattern)]
us_epi
```

```python
us_epi['location_key'] = us_epi['location_key'].str.replace('US_', '', regex=False)
us_epi
```

```python
#cleaning additional by setting the index
us_epi_cleaned = us_epi.set_index(['location_key'])
us_epi_cleaned
```

```python
#Exporting csv
us_epi_cleaned.to_csv('US_epi_cleaned.csv')
```
